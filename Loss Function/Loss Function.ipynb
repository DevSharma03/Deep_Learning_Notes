### Why is it Important?
- Without a loss function, the model would not know how to update weights.
- The loss function provides a signal for backpropagation (via gradients).
- Different tasks require different loss functions.

##### Classification Loss Function
1) We use **Log Loss** when the output of data is in the form of binary(0, 1) so we use this loss function.
2) When the results are in more than 2 types and less than 4 then we use **Categorical Cross Entropy**. It do one hot encoding and provide the result.
