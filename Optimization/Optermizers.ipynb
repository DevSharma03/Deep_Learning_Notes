# Optimizers (Gradient Decent We use Mostly)

An optimizer is an algorithm or method used to update the weights and biases of a neural network to minimize the loss function during training.

It decides how the model learns from data by adjusting parameters in the direction that reduces error.

Optimizers are algorithms or methods used to change the attribute of your neural network such as weights and learning rate in order to reduce the losses.

### Types of Optimizers

| Optimizer    | Pros                        | Cons                  | Best Used For     |
| ------------ | --------------------------- | --------------------- | ----------------- |
| **SGD**      | Simple, good generalization | Slow, oscillates      | Small datasets    |
| **Momentum** | Faster convergence          | Still oscillates      | CV tasks          |
| **NAG**      | Looks ahead, faster         | Complex math          | Deep CNNs         |
| **Adagrad**  | Handles sparse data         | Learning rate shrinks | NLP tasks         |
| **RMSprop**  | Works well in RNNs          | Sensitive to tuning   | Time-series       |
| **Adam**     | Fast, adaptive, robust      | May overfit sometimes | Most DL tasks     |
| **AdamW**    | Better generalization       | Slightly slower       | Transformers, NLP |
| **Nadam**    | Adam + Nesterov             | Costlier computation  | Advanced tasks    |
