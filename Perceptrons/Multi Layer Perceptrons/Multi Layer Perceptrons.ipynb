# Multi Layer Perceptrons (ANN)

### What is a Multi-Layer Perceptron (MLP)?
A Multi-Layer Perceptron (MLP) is an artificial neural network composed of multiple layers of nodes (neurons) beyond a single perceptron.
It is the simplest form of a feedforward neural network and is the foundation of deep learning.

- Input Layer → Takes features (X values).
- Hidden Layers → Perform nonlinear transformations of inputs using weights, biases, and activation functions.
- Output Layer → Produces predictions (classification or regression).

👉 Unlike a Single-Layer Perceptron (SLP) (which can only solve linearly separable problems), an MLP can learn non-linear decision boundaries.

### Artificial Neural Networks (ANN)
An Artificial Neural Network (ANN) is a computational model inspired by the structure and function of the human brain. Just like the brain has billions of interconnected neurons that process information, an ANN has artificial neurons (nodes) connected in layers that can learn to recognize complex patterns.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Sample Data (XOR Problem)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0, 1, 1, 0])   # XOR output

# ANN Model
model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))  # Hidden layer
model.add(Dense(1, activation='sigmoid'))            # Output layer

# Compile
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train
model.fit(X, y, epochs=200, verbose=0)

# Test
print("Predictions:", model.predict(X))
```
