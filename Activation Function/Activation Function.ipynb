# Activation Function

### What is an Activation Function?
An activation function is a mathematical function applied to the output of a neuron (weighted sum + bias) to decide whether the neuron should "activate" (fire) or not.

It introduces non-linearity into the neural network, allowing it to learn complex patterns beyond simple linear relationships.
