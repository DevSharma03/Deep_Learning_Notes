# Vanishing Gradient Problem

### What is the Vanishing Gradient Problem?
- The vanishing gradient problem occurs when gradients (the partial derivatives used in backpropagation) become very small (close to zero) as they are propagated backward through many layers.
- The Vanishing Gradient problem is encountered when training artigicial neural networks with gradient based learning methods and backpropagation. In such methods, during each iteraltion of training each of the neural network's weihts receives an updateproportional to the partial derivative of the error function with respect to the current weight.
